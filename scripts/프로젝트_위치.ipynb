```python
def trans_geo(addr):
    try:
        geo_location = maps.geocode(addr)[0].get('geometry')
        lat = geo_location['location']['lat']
        lng =  geo_location['location']['lng']
        return [lat,lng]
    except:
        return [0,0]
```

```python
for idx, addr in enumerate(tqdm(df_store.addr)):
    df_store.loc[idx,'latitude'] = trans_geo(addr)[0]
    df_store.loc[idx,'longitude'] = trans_geo(addr)[1]
```
